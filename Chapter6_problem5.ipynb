```python
import numpy as np
import matplotlib.pyplot as plt

beta = 0.1

def swish(x):
  return x / (1 + np.exp(- beta * x))

def swish_prime(x):
  ebx = np.exp(- beta * x)
  return (1 + (1 + x * beta) * ebx) / (1 + ebx) ** 2

x = np.linspace(-2, 2, 100)

plt.plot(x, swish(x))
plt.plot(x, swish_prime(x))
plt.show()
```
